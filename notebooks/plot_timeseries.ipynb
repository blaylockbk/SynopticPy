# Plot Timeseries

```python
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import numpy as np
import cartopy.crs as ccrs

import sys
sys.path.append('../')

from synoptic.services import stations_timeseries
from synoptic.plots import plot_timeseries, plot_timeseries_wind, map_timeseries
```

```python
params = dict(stid='wbb,kmry,ukbkb,qhw,psrim,psink',
              vars='air_temp,relative_humidity,ozone_concentration,wind_speed,wind_direction',
              units='English',
              recent=timedelta(days=2), verbose='HIDE')
plot_timeseries(**params)
```

## Special Plot for Winds

```python
# Major Downslope Windstorm Event in Northern Utah.
df = stations_timeseries(stid='UTPKL',
                         start=datetime(2020,9,7),
                         end=datetime(2020,9,10),
                         units='speed|mph', verbose='HIDE')

plot_timeseries_wind(df)
```

## Plot the map of the station locations

```python
fig, ax = plt.subplots(1,1, figsize=[5,5],
                       subplot_kw=dict(projection=ccrs.LambertConformal()))

params = dict(radius='ANIC1,50', network='96,2', recent=120, vars='wind_speed')
map_timeseries(ax=ax, verbose='HIDE', **params)
```
